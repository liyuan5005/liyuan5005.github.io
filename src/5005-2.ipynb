```python
def score(df):
	def f(a,topic):
		for word in subtopics[topic]:
			if word in a:
				return '1'
			break
	score_list = []
	for topic in subtopics.keys():
		df[topic]=df['text_x'].apply(f,args=(topic,))
		print('Once Topic')
		df_1 = df[df[topic] == '1']
		score_list.append((topic,round(df_1['stars_y'].mean(),2)))
	#score_list.append(np.percentile(df_1['stars_y'], [25, 50, 75])[2])
	return df,score_list

def score(df,business_id):
	def f(a,topic):
		for word in subtopics[topic]:
			if word in a:
				return '1'
			break
	score_list = {'business_id':business_id}
	for topic in subtopics.keys():
		df[topic]=df['text_x'].apply(f,args=(topic,))
		#print('Once Topic')
		df_1 = df[df[topic] == '1']
		score_list[topic]=round(df_1['stars_y'].mean(),2)
	#score_list.append(np.percentile(df_1['stars_y'], [25, 50, 75])[2])
	return score_list

subtopics=dict(
    {'location': ['place','way','location','area','places','spot','parking','street', 'local', 'walk', 'drive', 'near', 'space', 'located', 'shop', 'mall'],
    'open':['time', 'minutes', 'times', 'hour', 'home', 'today', 'later'],
    'environment':['restaurant', 'table', 'atmosphere', 'dish', 'dishes', 'room','restaurants', 'decor', 'kitchen', 'ambiance', 'plates', 'glass'],
    'value':['price','worth','quality','prices','high','pay','expect','average','cheap','expensive','priced','pricey','value','overpriced','cost','low'],
    'taste':['sweet','spicy','garlic','chili','sauces','honey','flavor','salty'],
    'food':['food','beef','pork','shrimp','egg','noodles','tofu','dumplings','rice','soup','duck','sausage','bacon'],
    'survice':['service','friendly','server', 'servers','served','wait','ordered', 'order'],
    'feeling':['great','best','delicious','amazing','happy','worth','fantastic','incredible','yummy','wonderful','perfect']})
```
